```python
import re
def clean_diagnosis(diag):
    if not diag or pd.isna(diag) or diag.strip() == '':
        return ''
    diag = str(diag).strip().upper()
    parts = re.split(r'\s*[-]+\s*', diag)
    return parts[0].strip()

mcc_cc_list = mccandcclist_df['ICDCode'].apply(clean_diagnosis).unique().tolist()
```

```python
import re
from collections import defaultdict

def parse_sdx(code):
    code = str(code).strip().upper()
    
    # Case 1: Check if the code ends with '-MCC' or '-CC'
    mcc_match = re.search(r'-\s*MCC\s*$', code)
    cc_match = re.search(r'-\s*CC\s*$', code)
    
    if mcc_match:
        return code[:mcc_match.start()].strip()  # Return base code for -MCC
    elif cc_match:
        return code[:cc_match.start()].strip()  # Return base code for -CC
    
    # Case 2: Check if the code is in mccandcclist
    if code in mccandcclist:
        return code  # Include the code as-is
    
    # Case 3: If none of the above, return None
    return None
```

```python
import pandas as pd

# ---------------------------------------------------
# 1. ICD NORMALIZATION (CRITICAL)
# ---------------------------------------------------
def normalize_icd(code):
    """
    Normalizes ICD codes so formats match lookup table
    Example:
    E11.9  -> E119
    I50-9  -> I509
    E11 9  -> E119
    """
    if pd.isna(code):
        return ""

    return (
        str(code)
        .upper()
        .strip()
        .replace(".", "")
        .replace("-", "")
        .replace(" ", "")
    )


# ---------------------------------------------------
# 2. LOAD MCC / CC LOOKUP (ONCE)
# ---------------------------------------------------
mcc_df = pd.read_excel(
    r"C:\Users\Arunkumara\Downloads\MCCCCList (1).xlsx",
    usecols=["ICDCode", "MCCorCC"]
)

mcc_lookup = (
    mcc_df
    .assign(ICDCode=lambda d: d["ICDCode"].apply(normalize_icd))
    .set_index("ICDCode")["MCCorCC"]
    .to_dict()
)


# ---------------------------------------------------
# 3. OPTIMIZED SDX EXTRACTOR
# ---------------------------------------------------
def extract_sdx(row_list, lookup):
    """
    Input  : list / tuple of ICD codes
    Output : (DX_Type, DX_Set, MCC_List, CC_List, General_List)
    """

    if not row_list or pd.isna(row_list):
        return "No MCC or CC", (), (), (), ()

    mcc, cc, general = set(), set(), set()

    for code in row_list:
        code_norm = normalize_icd(code)
        if not code_norm:
            continue

        flag = lookup.get(code_norm)

        if flag == "MCC":
            mcc.add(code_norm)
        elif flag == "CC":
            cc.add(code_norm)
        else:
            general.add(code_norm)

    # Priority logic
    if mcc:
        dx_type = "MCC"
        dx_set = tuple(sorted(mcc))
    elif cc:
        dx_type = "CC"
        dx_set = tuple(sorted(cc))
    else:
        dx_type = "No MCC or CC"
        dx_set = tuple(sorted(general))

    return (
        dx_type,
        dx_set,
        tuple(sorted(mcc)),
        tuple(sorted(cc)),
        tuple(sorted(general))
    )


# ---------------------------------------------------
# 4. APPLY ONCE (FAST & SAFE)
# ---------------------------------------------------
df3[
    [
        "A_DX_Type",
        "A_DX_Set",
        "A_MCC_List",
        "A_CC_List",
        "A_General_List"
    ]
] = df3["A_DX_List"].apply(
    lambda x: pd.Series(extract_sdx(x, mcc_lookup))
)


# ---------------------------------------------------
# 5. FINAL BUSINESS SDX SET
# ---------------------------------------------------

df3['SDX_Set'] = df2.apply(
    lambda x: "No MCC or CC" if x['A_DX_Type'] == 'No MCC or CC' else x['A_DX_Set'],
    axis=1
)
```
